#### Name
Name is a unique value but is there any added information in it?

```python
df['Name'].head(10)
```

```python
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
pd.crosstab(df['Title'], df['Sex']).sort_values(by = ['female', 'male'], ascending=False)
```

Save only the common titles.

```python
df['Title'] = df['Title'].replace(['Mlle', 'Ms'], 'Miss')
df['Title'] = df['Title'].replace('Mme', 'Mrs')
df["Title"] = df['Title'].apply(lambda x: 'Rare' if x not in ['Miss', 'Mrs', 'Mr', 'Master'] else x)
```

```python
pd.crosstab(df['Title'], df['Sex']).sort_values(by = ['female', 'male'], ascending=False)
```

```python
sns.countplot(x='Title', hue='Survived', data=df, palette='viridis')
plt.title('Survival Rate by Title')
plt.xlabel('Title')
plt.ylabel('Number of Passengers')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
df['Title_cat'] = df['Title'].astype('category').cat.codes
df['Title_cat']
```

### New features
We can create a new feature for family size which combines Parch (parents/children's) and SibSp (siblings/spouses). This will enable us to drop Parch and SibSp from our datasets.

```python
df['Family_Size'] = df['SibSp'] + df['Parch'] 
df['Family_Size'].value_counts().sort_index()
```

We can create another feature called Is_Alone based on the family size. 

```python
df['Is_Alone'] = 0
df.loc[df['Family_Size']==0, 'Is_Alone'] = 1
df.groupby(['Family_Size','Is_Alone']).size()
```

```python
sns.countplot(x='Is_Alone', hue='Survived', data=df, palette='BuPu')
plt.title('Survival Rate by Title')
plt.xlabel('Is_Alone')
plt.ylabel('Number of Passengers')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
colordict = {0: 'red', 1: 'green'}
plt.figure(figsize=(20, 8))
scatter = plt.scatter(df.index, df['Family_Size'], c=df['Survived'].map(colordict),  edgecolors='black', linewidths=0.3)
plt.title('Family Size vs Passenger Index by Survival')
plt.xlabel('Passenger Index')
plt.ylabel('Family Size')
legend_elements = [plt.Line2D([0], [0], marker='o', color='w', label='Did not survive',
                              markerfacecolor='red', markersize=10),
                   plt.Line2D([0], [0], marker='o', color='w', label='Survived',
                              markerfacecolor='green', markersize=10)]

plt.legend(handles=legend_elements, title='Survival Status', loc='upper right')
plt.show()
```

### Discretization
- Equal-width partitioning
- Equal-depth partitioning
- reasoning partitioning 

#### Equal-width
$W = \frac{(B A)}{N}$

```python
N = 5
df['Age_bin'] = pd.cut(df['Age_2'], bins=N, labels=False, include_lowest=True, right=False)

df['Age_bin'].value_counts().sort_index()
```

What do you think about that partition?

```python
sns.histplot(data=df, x='Age_2', color='skyblue', stat="probability", discrete=True)
plt.title('Age distribution')
plt.xlabel('Age')
plt.ylabel('Probability')
plt.show()
```

```python
sns.histplot(data=df, x='Age_bin', color='skyblue', stat="probability", discrete=True)
plt.title('Age distribution by bins')
plt.xlabel('Age')
plt.ylabel('Probability')
plt.show()
```

### Equal-depth 
 by probability  

```python
N = 5
df['Age_bin2'] = pd.qcut(df['Age_2'], q=N, labels=False, duplicates='drop')

df['Age_bin2'].value_counts().sort_index()
```

```python
sns.histplot(data=df, x='Age_bin2', color='skyblue', stat="probability", discrete=True)
plt.title('Age distribution by bins')
plt.xlabel('Age')
plt.ylabel('Probability')
plt.show()
```

#### Reasoning partition
By age groups

```python
bins = [0, 12, 20, 30, 45, float('inf')]
df['Age_bin3'] = pd.cut(df['Age_2'], bins=bins, labels=False, include_lowest=True)

df['Age_bin3'].value_counts().sort_index()
```

```python
sns.histplot(data=df, x='Age_bin3', color='skyblue', stat="probability", discrete=True)
plt.title('Age distribution by bins')
plt.xlabel('Age')
plt.ylabel('Probability')
plt.show()
```

### Normalization
- Min-max 
- Z-score
- Decimal scalie 

It is important to notice that new data (e.g. test data) is normalized by the same scale.

#### Min-Max normalization

Change the data to be with values between 0 to 1.

```python
scaler = MinMaxScaler()
scaler.fit(pd.DataFrame(df['Fare']))
df['Fare_min_max'] = scaler.transform(pd.DataFrame(df['Fare']))
print(scaler.data_min_)
print(scaler.data_max_)
df[['Fare','Fare_min_max']].head()
```

```python
sns.displot(df['Fare_min_max'], color='skyblue')
sns.displot(df['Fare'], color='skyblue')
```

#### Z-score normalization

Change the data to be with 0 mean and variance 1.

```python
z_scaler = StandardScaler()
z_scaler.fit(pd.DataFrame(df['Age_2']))
df['Age_norm'] = z_scaler.transform(pd.DataFrame(df['Age_2']))
print(df['Age_norm'].head())
sns.displot(df['Age_norm'], color='skyblue')
```

### Outlier

Search for abnormal values i.e. those that are at the edges of the distribution, by checking an abnormality of 3 standard deviations.

```python
data_mean = df['Age_2'].mean()
data_std = df['Age_2'].std()
cut_off = data_std * 3
lower_bound = data_mean - cut_off
upper_bound = data_mean + cut_off
df.loc[(df['Age_2'] > upper_bound) | (df['Age_2'] < lower_bound)]
```

### Data reduction

#### PCA

Use only numerical columns and without missing value.

```python
df1 = df[['Survived', 'Pclass', 'SibSp', 'Parch', 'Sex_cat', 'Embarked_cat', 'Title_cat', 
       'Is_Alone', 'Fare_min_max', 'Age_norm']]
df1.head()
```

Apply PCA - fit and transform

```python
pca = PCA(n_components = 0.999, svd_solver = 'full') 
pca.fit(df1, y='Survived')
df2 = pd.DataFrame(pca.transform(df1))
```

```python
pca.explained_variance_ratio_
```

```python
colordict = {0: 'red', 1: 'blue'}
plt.scatter(df2[0], df2[1], c = df['Survived'].map(colordict))
```
